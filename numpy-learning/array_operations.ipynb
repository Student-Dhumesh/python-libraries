**ARRAY OPERATIONS**

```python
import numpy as np
```

Arithmetic Operations

Vector

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
print(x)
print(y)
```

```python
print(x + y)
```

```python
print(x - y)
```

```python
print(x * y)
```

```python
print(x / y)
```

```python
print(x // y)
```

```python
print(x ** y)
```

Matrix

```python
x = np.arange(1,10).reshape((3,3))
y = np.arange(9,0,-1).reshape((3,3))

print(x)
print(y)
```

```python
print(x + y)
```

```python
print(x - y)
```

```python
print(x * y)
```

```python
print(x / y)
```

```python
print(x // y)
```

```python

print(x ** y)
```

**Deep and Shallow copy**

Shallow copy

```python
a = np.array([1,2,3,4,5])
print(a)
```

```python
b = a 
print(b)
```

```python
b[0] = 9

print(a)
print(b)
```

Deep copy

```python
x = np.array([1,2,3,4,5])
print(x)
```

```python
y = a.copy()
print(y)
```

```python
y[0] = 9
print(x)
print(y)
```

```python
arr = np.arange(1,10)
print(arr)
```

```python
arr[0:6] * 5
print(arr)
```

Matirx Operations

```python
A = np.array([[1,2],[3,4]])
print(A)
```

```python
B = np.array([[5,6],[7,8]])
print(B)
```

```python
print(A @ B)
```

```python
print(A.dot(B))
```

```python
print(np.dot(A,B))
```

```python
print(np.matmul(A,B))
```

```python
print(A.__matmul__(B))
```

Transpose

```python
arr = np.arange(1,10).reshape((3,3))
print(arr)
```

```python
print(arr.T)
```

```python
print(arr.transpose())
```

**Advance Array Manipulation**

Stacking Array

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8,])
print(a)
print(b)
```

```python
print(np.vstack((a,b)))
```

```python
print(np.hstack((a,b)))
```

```python
print(np.column_stack((a,b)))
```

Splitting Array

```python
arr = np.arange(1,17).reshape(4,4)
print(arr)
```

```python
np.hsplit(arr, 2)
```

```python
a = np.hsplit(arr, 4)
```

```python
a[0]
```

```python
for i in a:
    print(i)
```

```python
np.vsplit(arr, 2)
```

```python
a = np.vsplit(arr, 4)
```

```python
for i in a:
    print(i)
```
