```python
import numpy as np
```

**NUMPY INDEXING AND SLICING OF VECTORS**

```python
arr = np.arange(11,21)
print(arr)
```

```python
print(arr[1])
print(arr[5])
```

```python
print(arr[:6])
print(arr[6:])
```

```python
print(arr[1:6])
```

```python
print(arr[0::1])
print(arr[0::2])
print(arr[0::3])
```

**NUMPY INDEXING AND SLICING OF MATRIX**

```python
arr = np.arange(1,31).reshape((6,5))
print(arr)
```

```python
print(arr[0,1])
print(arr[3,0])
print(arr[4,1])
```

```python
print(arr[:3])
```

```python
print(arr[3:])
```

```python
print(arr[:,:3])
```

```python
print(arr[:,3:])
```

```python
print(arr[2:, 2:])
```

```python
print(arr[:4,:4])
```

**BOOLEAN INDEXING**

```python
arr = np.arange(11,21)
print(arr)
```

```python
even_arr = arr % 2 == 0
print(even_arr)
```

```python
even_arr = arr[even_arr]
print(even_arr)
```
